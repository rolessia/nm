Метод оберненої матриці +  Solve()

image.png

```python
import numpy as np

a = np.array([
    [4.24, 2.73, -1.55],
    [2.34, 1.27, 3.15],
    [3.05, -1.05, -0.63]])

b = np.array([[1.24], [2.35], [-1.25]])

det_a = np.linalg.det(a)
if det_a == 0:
    print("Визначник дорівнює нулю - розв'язків немає")
else:
  a_inv = np.linalg.inv(a)
  print(f"Обернена матриця до матриці A: \r\n {a_inv }")
  print(f"Вектор рішення: \r\n {a_inv.dot(b)}")

```

перевірка  Solve()

```python
print("Функція Solve()")
x_solve = np.linalg.solve(a, b)
print(f"Вектор рішення: \n {x_solve}")
```
