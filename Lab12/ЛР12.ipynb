```python
from scipy import integrate
import numpy as np

a, b, n = 1.2, 2, 10

h = (b - a) / n

def f1(x):
    return 1 / np.sqrt(0.5 * x + 1.5)

def left_rec(f1, a, b):
    sum = 0
    for i in range(0, n):
        sum += f1(a + i * h)
    return sum * h

print("Left rectangle:", round(left_rec(f1, a, b), 5))

def right_rec(f1, a, b):
    sum = 0
    for i in range(1, n+1):
        sum += f1(a + i * h)
    return sum * h

print("Right rectangle:", round(right_rec(f1, a, b), 5))

def aver_rec(f1, a, b):
    sum = 0
    for i in range(0, n):
        sum += f1(a + (i + 0.5) * h)
    return sum * h

print("Average rectangle:", round(aver_rec(f1, a, b), 5))

v, err = integrate.quad(f1, a, b)
print("Check with SciPy quad method:", round(v, 5))
```

```python
def f(x):
    return np.log10(x**2 + 3) / (2 * x)
a, b, n = 1.2, 2, 8
h = (b - a) / n

def simpson_rule(f, a, b, n):
    integr = f(a) + f(b)
    for i in range(1, n):
        k = a + i * h
        if i % 2 == 0:
            integr += 2 * f(k)
        else:
            integr += 4 * f(k)
    integr *= h / 3
    return integr

print("Simpsone method:", round(simpson_rule(f, a, b, n), 3))

v, err = integrate.quad(f, a, b)
print("Check for the simpsone method:", round(v, 3))
```

```python
def f(x):
    return 1 / np.sqrt(x**2 + 0.6)
a, b, n = 1.2, 2.6, 20

def trapezoidal_rule(f, a, b, n):
    h = (b - a) / n
    sum = 0.5 * (f(a) + f(b))
    x = a

    for i in range(1, n):
        x += h
        sum += f(x)

    integral = h * sum

    return integral

integral = trapezoidal_rule(f, a, b, n)
print("Trapetzia method result:", round(integral, 5))

v, err = integrate.quad(f, a, b)
print("Check with SciPy quad method:", round(v, 5))
```
