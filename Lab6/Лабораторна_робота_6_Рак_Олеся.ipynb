```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, expand
from scipy.interpolate import lagrange

x = np.array([-3., -1., 1., 2.], dtype=float)
y = np.array([3., 3., -13., -12.], dtype=float)
x1 = -4.0
x2 = -2.0
x3 = -1.5
x4 = 0.5

def lagrange_interpolation(x, y, x_test):
    n = len(x)
    p = np.zeros(n) # масив для зберігання значень багаточленів L_i
    for i in range(n):
        # обчислюємо багаточлен L_i
        p_i = 1
        for j in range(n):
            if i != j:
                p_i *= (x_test - x[j]) / (x[i] - x[j])
        p[i] = p_i
    return np.dot(y, p)

f1 = lagrange_interpolation(x, y, x1)
print(f"Значення функції у точці x_1 = ", f1.round(4))

f2 = lagrange_interpolation(x, y, x2)
print(f"Значення функції у точці x_2 = ", f2.round(4))

f3 = lagrange_interpolation(x, y, x3)
print(f"Значення функції у точці x_3 = ", f3.round(4))

f4 = lagrange_interpolation(x, y, x4)
print(f"Значення функції у точці x_4 = ", f4.round(4))
```

```python
xnew = np.linspace(min(x.min(), x1), max(x.max(), x4, 2), 100)
ynew = [lagrange_interpolation(x, y, i) for i in xnew]

plt.plot(x,y,'o',xnew, ynew)
plt.plot(x1,f1,'go')
plt.plot(x2,f2,'go')
plt.plot(x3,f3,'go')
plt.plot(x4,f4,'go')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Lagrange Polynomial_1')
plt.grid(True)
plt.show()
```

```python
f = lagrange(x, y)
fig = plt.figure(figsize = (7,5))
plt.plot(x, y, 'o', xnew, f(xnew))
plt.title('Lagrange Polynomial_2')

plt.plot(x1,f1,'ro')
plt.plot(x2,f2,'ro')
plt.plot(x3,f3,'ro')
plt.plot(x4,f4,'ro')

plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.array([3, 4, 5, 6], dtype=float)

def f_x(x):
    return 3 / (x + 1)

y_data = f_x(x)

print('x =', x)
print('f_x(x) =', np.round(y_data, 5))
print('shapes:', x.shape, y_data.shape)

p = lagrange(x, y_data)
print('\nПоліном Лагранжа p(x):')
print(p)
print('Ступінь полінома:', p.order)

z = np.linspace(np.min(x), np.max(x), 100)

plt.figure(figsize=(8, 5))
plt.plot(z, p(z), '-', label='$L(x)$')
plt.plot(x, y_data, 'o', label='data')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
