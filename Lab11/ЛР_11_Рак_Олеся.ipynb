```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return np.sin(2 * x) + x

x = np.array([i * 0.1 for i in range(1, 11)])
y = np.array([func(xi) for xi in x])

print('x =', np.round(x, 2))
print('y =', np.round(y, 4))
print('-' * 40)

n = len(x)
x_mean = np.mean(x)
y_mean = np.mean(y)

a = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - np.sum(x)**2)
b = y_mean - a * x_mean

print(f"Рівняння прямої: y = {a:.4f}x + {b:.4f}")

coeffs_parabola = np.polyfit(x, y, 2)
a2, b2, c2 = coeffs_parabola

print(f"Рівняння параболи: y = {a2:.4f}x^2 + {b2:.4f}x + {c2:.4f}")
print('-' * 40)

plt.figure(figsize=(10, 6))

plt.scatter(x, y, color='red', zorder=5, label='Точки даних f(x)')

x_plot = np.linspace(0.05, 1.05, 100)
y_line_plot = a * x_plot + b
y_parabola_plot = a2 * x_plot**2 + b2 * x_plot + c2

plt.plot(x_plot, y_line_plot, color='blue', linestyle='--', label='Наближення прямою')
plt.plot(x_plot, y_parabola_plot, color='green', label='Наближення параболою')

plt.xlabel('x')
plt.ylabel('y')
plt.title('МНК. Наближення прямою та параболою')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return np.sin(2 * x) + x

x = np.array([i * 0.1 for i in range(1, 11)])
y = np.array([func(xi) for xi in x])

print('x=', x)
print('y=', y)

coefficients = np.polyfit(x, y, 2)

a, b, c = coefficients
print(f"Рівняння параболи: y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}")

polynomial = np.poly1d(coefficients)

plt.scatter(x, y, color='red', label='Точки даних')
x_line = np.linspace(min(x), max(x), 100)
plt.plot(x_line, polynomial(x_line), color='blue', label='Апроксимація поліномом')

plt.title('Поліном другого порядку МНК за допомогою polyfit()')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import least_squares

def func(x):
    return np.sin(2 * x) + x

x = np.array([i * 0.1 for i in range(1, 11)])
y = np.array([func(xi) for xi in x])

print('x=', np.round(x, 2))
print('y=', np.round(y, 4))

def fun(a, x, y):
    return a[0] + a[1] * x + a[2] * x**2 - y

a0 = np.array([1, 1, 1])

res_lsq = least_squares(fun, x0=a0, args=(x, y))

print("a0 = %.2f, a1 = %.2f, a2 = %.2f" % tuple(res_lsq.x))

def f(x):
    return res_lsq.x[0] + res_lsq.x[1] * x + res_lsq.x[2] * x**2

x_p = np.linspace(min(x), max(x), 100)
y_p = f(x_p)

plt.plot(x, y, 'o', color='red', label='Дані')
plt.plot(x_p, y_p, 'b', label='МНК апроксимація параболою')
plt.title("МНК наближення параболою")
plt.legend()
plt.grid(True)
plt.show()
```
