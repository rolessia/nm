```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_min, x_max = -2, 2
y_min, y_max = -2, 2
step = 0.01
```

```python
x, y = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
```

```python
eq1 = 2 * x - np.cos(y + 1)
eq2 = y + np.sin(x) + 0.4
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
ax.contour(x, y, eq1, levels=[0], colors='red')
ax.contour(x, y, eq2, levels=[0], colors='blue')
```

```python
ax.set_xlim([x_min, x_max])
ax.set_ylim([y_min, y_max])
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Графік системи рівнянь')
fig
```

```python
ax.arrow(x_min, 0, x_max - x_min, 0, head_width=0.1, head_length=0.05, fc='green', ec='green')
ax.arrow(0, y_min, 0, y_max - y_min, head_width=0.1, head_length=0.2, fc='green', ec='green')
fig
```

```python
import numpy as np
from scipy import optimize

x0, y0 = 0.4, -0.8

def f1(y):
    return 0.5 * np.cos(y + 1)

def f2(x):
    return -0.4 - np.sin(x)
def iter(x, y, e):
    xn = x
    yn = y
    xn1 = f1(yn)
    yn1 = f2(xn)
    n = 1

    while (abs(xn1 - xn) >= e) or (abs(yn1 - yn) >= e):
        xn = xn1
        yn = yn1
        xn1 = f1(yn)
        yn1 = f2(xn)
        n += 1

    print('Simple iteration:')
    print('x =', xn1, '\ny =', yn1, '\nThe amount of iteration =', n)
iter(x0, y0, 0.001)

def f3(vars):
    return 2 * vars[0] - np.cos(vars[1] + 1), vars[1] + np.sin(vars[0]) + 0.4
s = optimize.root(f3, [0.4, -0.8], method='hybr')
print('Chek', s.x)
```
